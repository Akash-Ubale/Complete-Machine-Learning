# Eclat

Apriori Model is recommended.

it's a simplified version of the Apriori model. 

It's basically the Apriori model without Confidence and Lift.

In Apriori we had 3 components Support, Confidence, Lift, But in Eclat we have only Support, which is calculated for a set of entities.

taking the same Movie recommendation example.
$$Support(M) = \frac{No.~ of~ User~ Watch~ lists~ containing~ M}{Total~No.~ of~ User~ Watch~ lists~}$$
$M - A~specific~set~of~Movies $

### Algorithm:
1. Set minimum Support
2. take all the usbset transactions having support higher than the min support.
3. Sort these subsets by decresing support values.

### problem statement:
Same as Apriori - The owner wants to create a deal 'Buy one get one free'. We need to find the suitable products for that, which gives best sales and profits.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Preprocessing

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)

# As the Eclat model needs the data in list format.
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
dataset.head()
```

## Training the Eclat model on the dataset

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, 
                min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```python
results = list(rules)
results
```

### Putting the results well organised into a Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

### Displaying the results sorted by descending supports

```python
resultsinDataFrame.nlargest(n = 10, columns = 'Support')
```
